import pandas as pd
import matplotlib.pyplot as plt


# Step 1: Load data from the CSV file


```python
file_name = "online_sales.csv"
df = pd.read_csv("C:/Users/mahes/Downloads/online_sales.csv")
```

# Step 2: Aggregate sales data


```python
summary = df.groupby("product").agg(
    total_boxes=("boxes_shipped", "sum"),
    revenue=("amount", "sum")
).reset_index()
```

# Step 3: Print the summary


```python

print("Sales Summary:")
print(summary)
```

# Step 4: Visualize revenue by product


```python
plt.figure(figsize=(8, 6))
plt.bar(summary["product"], summary["revenue"], color="orange")
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.savefig("sales_chart.png")
plt.show()
```

# Step 5: Export the summary to a CSV file (optional)


```python
summary.to_csv("sales_summary.csv", index=False)
```
